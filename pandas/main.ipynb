```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("./data/raw/store.csv")

x = df['region']
y = df['sales']

plt.bar(x, y)

plt.title("bar region", fontsize=16, fontweight="bold", color="red", fontstyle="italic")
plt.xlabel("region", fontsize=16, fontweight="bold")
plt.ylabel("sales", fontsize=10, fontweight="bold")

plt.grid(color='', linestyle="dashed", linewidth=1, alpha=0.5)

plt.show()   
```
